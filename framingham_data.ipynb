```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Dataset1/framingham.csv')
df
```

```python
df.isna().sum()
```

```python
df['education'].value_counts()
```

```python
# md=df['education'].mode()[0]
# df['education']=df['education'].fillna(md)

df['education']=df['education'].fillna(df['education'].mode()[0])
```

```python
df
```

```python
df['education'].value_counts()
```

```python
df['cigsPerDay']=df['cigsPerDay'].fillna(df['cigsPerDay'].mode()[0])
df
```

```python
df['BPMeds'].value_counts()
```

```python
df['BPMeds']=df['BPMeds'].fillna(df['BPMeds'].mode()[0])
df
```

```python
df['totChol'].value_counts()
```

```python
df['totChol']=df['totChol'].fillna(df['totChol'].mode()[0])
df
```

```python
df['BMI'].value_counts()
```

```python
df['BMI']=df['BMI'].fillna(df['BMI'].mean())
df
```

```python
df['heartRate'].value_counts()
```

```python
df['heartRate']=df['heartRate'].fillna(df['heartRate'].mode()[0])
df
```

```python
df['glucose'].value_counts()
```

```python
df['glucose']=df['glucose'].fillna(df['glucose'].mode()[0])
df
```

```python
df.drop(['education'],axis=1,inplace=True)
df
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=1)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.svm import SVC
sv=SVC()
sv.fit(X_train,y_train)
```

```python
y_pred=sv.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,classification_report
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
```

```python
y_pred1=knn.predict(X_test)
y_pred1
```

```python
print(accuracy_score(y_test,y_pred1))
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred1))
print(classification_report(y_test,y_pred1))
```
